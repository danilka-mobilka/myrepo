```python
import time
import matplotlib.pyplot as plt
import random

# Define insertion sort
def insertion_sort(arr):
    for j in range(1, len(arr)):
        key = arr[j]
        i = j - 1
        while i >= 0 and arr[i] > key:
            arr[i + 1] = arr[i]
            i -= 1
        arr[i + 1] = key

# Define bubble sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

# Sizes for testing
n_values = [5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000, 10000]

# Store times for each algorithm
insertion_times = []
bubble_times = []

# Measure execution time for each array size
for n in n_values:
    # Generate a random array of size n
    test_array = [random.randint(0, 100000) for _ in range(n)]

    # Measure time for insertion sort
    start_time = time.time()
    insertion_sort(test_array.copy())  # use a copy to avoid modifying the original
    insertion_times.append(time.time() - start_time)

    # Measure time for bubble sort (only for smaller arrays to prevent long execution)
    if n <= 5000:  # Restrict bubble sort to smaller arrays
        start_time = time.time()
        bubble_sort(test_array.copy())  # use a copy to avoid modifying the original
        bubble_times.append(time.time() - start_time)
    else:
        bubble_times.append(None)  # Placeholder for large sizes

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(n_values, insertion_times, label="Insertion Sort", marker='o')
plt.plot(n_values[:len(bubble_times)], bubble_times, label="Bubble Sort", marker='o')
plt.xlabel("Array Size (n)")
plt.ylabel("Execution Time (seconds)")
plt.title("Execution Time of Insertion Sort and Bubble Sort")
plt.yscale("log")  # Logarithmic scale for better visualization
plt.legend()
plt.grid(True)
plt.show()
```
